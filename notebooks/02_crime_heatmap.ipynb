```python
import pandas as pd
import folium
```

```python
df = pd.read_csv(
    "../data/raw/ksp_synthetic_crime_dataset.csv"
)

df.head()
```

```python
crime_map = folium.Map(
    location=[13.0, 76.0],
    zoom_start=7
)

crime_map
```

```python
for _, row in df.head(500).iterrows():

    folium.CircleMarker(
        location=[
            row["latitude"],
            row["longitude"]
        ],

        radius=4,

        popup=f"""
        District: {row['district']}
        Crime: {row['crime_type']}
        """,

        fill=True

    ).add_to(crime_map)
```

```python
crime_map
```

```python
crime_map.save(
    "../outputs/maps/crime_map.html"
)
```
