```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
crime_map = folium.Map(
    location=[13.0, 76.0],
    zoom_start=7
)
```

```python
df = pd.read_csv(
    "../data/raw/ksp_synthetic_crime_dataset.csv"
)

df.head()
```

```python
heat_data = df[
    ["latitude", "longitude"]
].values.tolist()
```

```python
HeatMap(
    heat_data,
    radius=12,
    blur=15
).add_to(crime_map)
```

```python
crime_map
```

```python
crime_map.save(
    "../outputs/maps/crime_heatmap.html"
)
```
